Arbre de decission

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# 1. Charger les données
df = pd.read_csv('./golf_df (1).csv')

# 2. Séparer la cible (Play) des variables explicatives
y = df['Play']
X = df.drop(columns='Play')

# 3. Encodage des variables
#    - Pour la cible, un LabelEncoder suffira (yes/no → 1/0)
le_y = LabelEncoder()
y_encoded = le_y.fit_transform(y)

#    - Pour les entrées catégorielles, on peut créer des variables indicatrices (one‑hot)
X_encoded = pd.get_dummies(X, drop_first=True)

# 4. Diviser en jeu d’entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_encoded, test_size=0.3, random_state=42
)

# 5. Entraîner l’arbre de décision
tree = DecisionTreeClassifier(
    criterion='entropy',  
    max_depth=None,
    random_state=42
)
tree.fit(X_train, y_train)

# 6. Évaluer le modèle
y_pred = tree.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Exactitude de l'arbre sur le jeu de test : {accuracy:.2f}")

# 7. les règles de décision de l’arbre
print("\nRègles extraites de l'arbre :")
rules_text = export_text(tree, feature_names=list(X_encoded.columns))
print(rules_text)
```
